```python
import matplotlib.pyplot as plt
import numpy as np 
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import pandas as pd
```

```python
transform = transforms.ToTensor()

# Tải tập huấn luyện (Train)
print("Đang tải tập Train...")
train_dataset = datasets.MNIST(root='../data',     
                               train=True,          
                               download=True,      
                               transform=transform)
# Tải tập kiểm tra (Test)
print("Đang tải tập Test...")
test_dataset = datasets.MNIST(root='../data', 
                              train=False,    
                              download=True, 
                              transform=transform)
print(f"Hoàn tất! Số lượng ảnh train: {len(train_dataset)}")
print(f"Hoàn tất! Số lượng ảnh test: {len(test_dataset)}")
print(f"Kích thước ảnh train: {train_dataset.data.shape}")
print(f"Kích thước ảnh test: {test_dataset.data.shape}")
```

```python
np.random.seed(42)
projection_matrix = np.random.rand(28 * 28, 128) 
def preprocess(dataset):
    X = dataset.data.numpy()
    y = dataset.targets.numpy()
    mask = (y == 1) | (y == 7)
    X_filtered = X[mask]
    y_filtered = y[mask]
    X_flattened = X_filtered.reshape(-1, 28 * 28)
    X_flattened = X_flattened / 255.0
    X_projected = np.dot(X_flattened, projection_matrix)
    return X_projected, y_filtered
train_X, train_y = preprocess(train_dataset)
test_X, test_y = preprocess(test_dataset)
print(f"Kích thước sau khi chiếu: {train_X.shape}") 
```

```python
logreg = linear_model.LogisticRegression(max_iter=1000)
print("Đang huấn luyện mô hình...")
logreg.fit(train_X, train_y)
test_predictions = logreg.predict(test_X)
accuracy = logreg.score(test_X, test_y)
print(f"Độ chính xác trên tập test: {accuracy * 100:.2f}%")
```
