```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
len(iris.data)
```

```python
clusters = len(pd.DataFrame(iris_y)[0].unique()) # Sẽ bằng 3
kmean = KMeans(n_clusters=clusters, random_state=1, n_init=10)
kmean.fit(iris_X)
labels = kmean.labels_
centers = kmean.cluster_centers_

plt.figure(figsize=(10, 6))
plt.scatter(iris_X[:, 0], iris_X[:, 1], c=labels, s=50, cmap='viridis', alpha=0.6, label='Data points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("Kết quả phân cụm bằng K-Means")
plt.xlabel(iris.feature_names[0]) 
plt.ylabel(iris.feature_names[1]) 
plt.legend()
plt.show()
```

```python
test_size = len(iris_X) // 4
X_train, X_test, y_train, y_test = train_test_split(
     iris_X, iris_y, test_size=test_size, random_state=1)
print(f"Training size: {len(X_train)}")
print(f"Test size: {len(X_test)}")
```

# Training model

```python
n_neighbors = 6
model = neighbors.KNeighborsClassifier(n_neighbors = n_neighbors, weights="distance")
model.fit(X_train, y_train)
```

# Test model

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_test)
print(f"Accuracy(%): {accuracy_score(y_pred=y_pred, y_true=y_test) * 100}%")
```
